```python
import seaborn as sns
import pandas as pd

dinodata = pd.read_csv('DatasaurusDozen.tsv', sep='\t') #load data into a pandas object
                                                        #specifying separation value as a tab 
```

```python
sns.lmplot(data = dinodata,
           x='x',
           y='y',
           col = 'dataset',
           hue = 'dataset',
           line_kws={'color':'#000000'},
           col_wrap = 4,
           height = 3,
           markers='o'
           # ci = None # skips bootstrapping to save computational budget
          )
```

```python
anscombe = sns.load_dataset("anscombe")
sns.lmplot(data = anscombe,
           x='x',
           y='y',
           scatter_kws={'s':80},
           col = 'dataset',
           col_wrap = 2,
           ci = None        
          )
```
